```python
import matplotlib.pyplot as plt
import numpy as np

# Training data
hours_x = [1.0, 2.0, 3.0]
points_y = [2.0, 4.0, 6.0]

if __name__ == "__main__":
    w_list = []
    mse_list = []
    for w in np.arange(0.0, 4.1, 0.1):
        sum = 0.0
        for x, y in zip(hours_x, points_y):
            predicted_y = w * x
            loss = abs(predicted_y - y)
            loss = loss * loss
            sum  = sum + loss
            print("x: {} y: {} loss : {}".format(x,y,loss))
        mse = sum / 3 #since no of training data is 3
        print("mse {}".format(mse))
        w_list.append(w)
        mse_list.append(mse)

    plt.plot(w_list,mse_list)
    plt.xlabel("w")
    plt.ylabel("loss")
    plt.show()
```
